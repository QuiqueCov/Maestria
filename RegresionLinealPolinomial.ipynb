```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values # variable independiente
y = dataset.iloc[:, -1].values # variable dependiente
```

```python
print(x)
```

```python
# Entrenamiendo del modelo de regression multiple
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
lin_regressor.fit(x, y) # esto es aprendizaje supervisado
```

```python
# Entrenamiento del modelo de regression polinomial
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly, y)
```

```python
print(x_poly)
```

```python
#Visualizar los resultados de la regresion lineal
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_regressor.predict(x), color = 'blue')
plt.title('Verdad o mentira (Regresion Lineal)')
plt.xlabel('Nivel de puesto')
plt.ylabel('Salario')
plt.show
```

```python
# visualizar la regrsion polinomial
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Verdad o mentira (Regresion Polinomial)')
plt.xlabel('Nivel de puesto')
plt.ylabel('Salario')
plt.show
```

```python
# Predecir un nuevo dato usando regresion lineal
print(lin_regressor.predict([[8.5]]))

# Nuevo dato usando polinomial
print(lin_reg2.predict(poly_reg.fit_transform([[8.5]])))
```
