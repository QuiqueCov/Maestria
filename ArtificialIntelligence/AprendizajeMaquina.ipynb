La base de datos Wine se utiliza principalmente para clasificar vinos en tres categorías (clases) basadas en una serie de características químicas. Es un conjunto de datos multidimensional, donde cada vino tiene un conjunto de atributos que describen sus características físicas y químicas.


# Caracteristicas:
Estas características son medidas a partir de un análisis químico de las muestras. A continuación se detallan los 13 atributos:

*  Alcohol: El contenido de alcohol en el vino, expresado como porcentaje
(porcentaje en volumen).

* Malic acid: Ácido málico, un compuesto orgánico que se encuentra en las uvas.

* Ash: Ceniza, un indicador de la cantidad de minerales presentes en el vino.

* Alcalinity of ash: Alcalinidad de la ceniza, relacionado con los compuestos alcalinos.

* Magnesium: Cantidad de magnesio en el vino.
* Total phenols: Fenoles totales, compuestos químicos relacionados con el sabor y color del vino.

* Flavanoids: Flavonoides, un tipo de compuesto fenólico con efectos antioxidantes.

* Nonflavanoid phenols: Fenoles no flavonoides, otra clase de compuestos fenólicos.

* Proanthocyanins: Proantocianinas, compuestos que afectan el color y la astringencia del vino.

* Color intensity: Intensidad del color del vino.

* Hue: El tono o matiz del color del vino.

* OD280/OD315 of diluted wines: Relación de absorbancia a 280 nm y 315 nm, un indicador de la concentración de ciertos compuestos fenólicos.

* Proline: Un aminoácido que se encuentra en las uvas, y cuya concentración puede ser indicativa de ciertos tipos de vino.

##Bunch
Al querer imprimir los datos me encontre con este objeto llamado bunch,

El cual me di cuenta que tenia dos arreglos, lo cual hace que sea diferente a los ejemplos, donde era solo una tablita.

Contiene los arreglos data y target

Leyendo la documentacion de este set de datos me di cuenta de lo siguiente:

* wine.data:
Es una lista o arreglo que contiene las caracteristicas de cada vino (Alcohol, hue, ash, etc)

* wine.target:
Es una lista o arreglo que contiene las clases de cada vino entre, 0, 1 y 2.

Por lo tanto concluyo lo siguiente:

* wine.data:
Tiene las variables que influiran en la prediccion.

* wine.target:
Contiene la variable dependiente(la clase de vino), lo que quiero predecir.

```python
# Establecido lo anterior lo separe de la siguiente manera
X = wine.data  # caracteristicas
y = wine.target  # Clases
```

```python
#X
```

```python
#y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# Aqui lo que hago es la division de los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
'''
Aqui lo que hago es normalizar los datos, con el standar scaler,
ya que como los usare para todos los modelos es mejor usarlo desde aqui
'''
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#Regression Logistica
Solamente adapto los datos que ya tengo a la regresion logistica


```python
from sklearn.linear_model import LogisticRegression # importo el modelo de la regression logistica
log_regressor = LogisticRegression()
log_regressor.fit(X_train, y_train)
```

```python
# Prediccion de los resultados del conjunto de pruebas
y_pred = log_regressor.predict(X_test)
print(y_pred)
print(X_test)
print(y_test)
# aqui los imprimo pero la verdad es demasiada informacion
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred) # el r2 score para ver la exactitud del modelo
```

# Prediccion
Esta linea de codigo es el vino que quiero predecir, se reutilizara para todos los modelos


```python
# Creo el nuevo vino
nuevo_vino = np.array([13.2, 1.78, 2.14, 3.12, 105, 2.54, 2.4, 0.3, 1.0, 0.3, 1.3, 3.25, 1100]).reshape(1, -1)

# Normalizo los datos para evitar cualquier error
nuevo_vino = scaler.transform(nuevo_vino)
```

```python
# Hago la prediccion
prediccion_log_reg = log_regressor.predict(nuevo_vino)
print(f"Clase del vino segun Logic Regression: {prediccion_log_reg[0]}")
```

# K-Nearest Neigbors
Solamente adapto los datos a este modelo


```python
# Crear y entrenar el modelo de K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Hago la prediccion
prediccion_knn = knn.predict(nuevo_vino)
print(f"Clase de vino segun  K-NN: {prediccion_knn[0]}")
```

##Exactitud

```python
from sklearn.metrics import accuracy_score

# Realizar predicciones en el conjunto de prueba
y_pred_knn = knn.predict(X_test)

# Calcular la exactitud
exactitud_knn = accuracy_score(y_test, y_pred_knn)

# Mostrar el resultado
print(f"Exactitud del modelo K-NN: {exactitud_knn:.4f}")
```

#Support Vector Machine(SVM)
Solamente adapto los datos a este Modelo

```python
# Crear y entrenar el modelo de Support Vector Machine
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
```

```python
# Hago la prediccion
prediccion_svm = svm.predict(nuevo_vino)
print(f"Clase de vino segun SVM: {prediccion_svm[0]}")
```

##Exactitud

```python
from sklearn.metrics import accuracy_score

# Realizar predicciones en el conjunto de prueba
y_pred_svm = svm.predict(X_test)

# Calcular la exactitud
exactitud_svm = accuracy_score(y_test, y_pred_svm)

# Mostrar el resultado
print(f"Exactitud del modelo SVM: {exactitud_svm:.4f}")
```

# Decision Tree
Solamente adapto los datos a este modelo

```python
# Crear y entrenar el modelo Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
# Hago La prediccion
prediccion_tree = tree.predict(nuevo_vino)
print(f"Clase de Vino Segun Decission Tree: {prediccion_tree[0]}")
```

## Exactitud

```python
# Realizar predicciones en el conjunto de prueba
y_pred_tree = tree.predict(X_test)

# Calcular la exactitud
exactitud_tree = accuracy_score(y_test, y_pred_tree)

# Mostrar el resultado
print(f"Exactitud de Decision Tree: {exactitud_tree:.4f}")
```

#Random Forest}
Solamente adapto los datos a este modelo


```python
# Crear y entrenar el modelo de Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
#prediccion
prediccion_rf = rf.predict(nuevo_vino)
print(f"Clase de Vino segun Random Forest es: {prediccion_rf[0]}")
```

## Exactitud

```python
# Realizar predicciones en el conjunto de prueba
y_pred_rf = rf.predict(X_test)

# Calcular la exactitud
exactitud_rf = accuracy_score(y_test, y_pred_rf)
print(f"Exactitud de Random Forest: {exactitud_rf:.4f}")
```
